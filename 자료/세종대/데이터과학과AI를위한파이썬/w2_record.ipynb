### 3. 그래프 그리기
#### 1) 함수 x^2 그래프
* numpy.linspace(start, stop, num)
* matplotlib.pyplot(맷플롯립)

```python
def f(x):
    return (x-2)**2 + 2
```

```python
import numpy as np
x= np.linspace(-1, 5, 101)
```

```python
x
```

```python
y = f(x)
```

```python
y
```

```python
# matplotlib 한글 폰트 사용
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/gulim.ttc"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)
```

```python
import matplotlib.pyplot as plt
plt.plot(x, y, "b-")
#plt.xlim(-2, 2)
#plt.ylim(0, 5)
plt.title("함수 $f(x) = x^2$ 의 그래프")
plt.xlabel("X")
plt.ylabel("f(x)")
plt.rcParams['figure.figsize'] = [10, 10]
plt.grid(True)
plt.show()
```

#### 2) 지수함수 그래프 f(x) = e^x
* numpy.linspace(start, stop, num)
* matplotlib.pyplot(맷플롯립)

```python
xx = np.linspace(-2, 2, 100)
yy = np.exp(xx)
plt.plot(xx, yy)
plt.title("지수함수($e^x$)의 그래프")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

#### 3) 로그함수 그래프 f(x) = ln(x)
* numpy.linspace(start, stop, num)
* matplotlib.pyplot(맷플롯립)

```python
xx = np.linspace(0.01, 8, 100)
yy = np.log(xx)
plt.plot(xx, yy)
plt.grid(True)
plt.axhline(0, c='r', ls = '--')
plt.axvline(0, c='r', ls = '--')
plt.show()
```

#### 4) 로지스틱함수 그래프
* numpy.linspace(start, stop, num)
* matplotlib.pyplot(맷플롯립)

```python
def logistic(x):
    return 1/(1+np.exp(-x))
xx = np.linspace(-10, 10, 100)
plt.plot(xx, logistic(xx))
plt.grid(True)
plt.show()
```
